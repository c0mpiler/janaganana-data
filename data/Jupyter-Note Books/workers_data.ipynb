```python
file_path="/home/arun/workspace/projects/janaganana-data/data/workers_data/B-01/csv/workers_data.csv"

# Reading csv file for pca data
import pandas as pd
df = pd.read_csv(file_path)
```

```python
df.head(4)
```

```python
# only interested in literacy of all ages
# if age specific is required then additional conditions need to be added here
df = df[((df.area_type == "Rural") | (df.area_type == "Urban")) & (df.age_group == "Total")]
```

```python
# pick required columns
df=df[['state_code', 'district_code', 'area_type', 'area_name', u'main_workers_m', u'main_workers_f', u'marginal_less3_m', u'marginal_less3_f', u'marginal_3to6_m', u'marginal_3to6_f', u'marginal_available_m', u'marginal_available_f', u'nonworkers_m', u'nonworkers_f', u'nonworkers_available_m', u'nonworkers_available_f']]
```

```python
def func(ds):
    ds.area_name = ds.area_name.lower();
    if ds.area_name.startswith('district'):
        ds['geo_code'] = ds.district_code;
    else:
        ds['geo_code'] = ds.state_code;
        
    if ds.area_name == 'india':
        ds['geo_level'] = 'country'
        ds['geo_code'] = 'IN'
    elif ds.area_name.startswith('district'):
        ds['geo_level'] = 'district'
    elif ds.area_name.startswith('state'):
        ds['geo_level'] = 'state'   
        
    return ds

df=df.apply(func, axis=1)
```

```python
df.head(4)
```

```python
df = df[['geo_code', 'geo_level', 'area_type', u'main_workers_m', u'main_workers_f', u'marginal_less3_m', u'marginal_less3_f', u'marginal_3to6_m', u'marginal_3to6_f', u'marginal_available_m', u'marginal_available_f', u'nonworkers_m', u'nonworkers_f', u'nonworkers_available_m', u'nonworkers_available_f']]
df.head(4)
```

```python
# isloate 'Male' data
df1 = df[['geo_code', 'geo_level', 'area_type', u'main_workers_m', u'marginal_less3_m', u'marginal_3to6_m', u'marginal_available_m', u'nonworkers_m', u'nonworkers_available_m']]
df1['Sex'] = 'Male';
df1.rename(columns={'main_workers_m':'Main workers'}, inplace=True)
df1.rename(columns={'marginal_less3_m':'Marginal workers worked less than 3 months'}, inplace=True)
df1.rename(columns={'marginal_3to6_m':'Marginal workers worked 3 to 6 months'}, inplace=True)
df1.rename(columns={'marginal_available_m':'Marginal workers available for work'}, inplace=True)
df1.rename(columns={'nonworkers_m':'Non-workers'}, inplace=True)
df1.rename(columns={'nonworkers_available_m':'Non-workers available for work'}, inplace=True)

# Extract each education level and add rows corresponding to each value
dmale_level = pd.melt(df1, id_vars=['geo_code', 'geo_level', 'area_type', 'Sex'], var_name="Workers category", value_name="Category total")
```

```python
# isloate 'Female' data
df2 = df[['geo_code', 'geo_level', 'area_type', u'main_workers_f', u'marginal_less3_f', u'marginal_3to6_f', u'marginal_available_f', u'nonworkers_f', u'nonworkers_available_f']]
df2['Sex'] = 'Female';
df2.rename(columns={'main_workers_f':'Main workers'}, inplace=True)
df2.rename(columns={'marginal_less3_f':'Marginal workers worked less than 3 months'}, inplace=True)
df2.rename(columns={'marginal_3to6_f':'Marginal workers worked 3 to 6 months'}, inplace=True)
df2.rename(columns={'marginal_available_f':'Marginal workers available for work'}, inplace=True)
df2.rename(columns={'nonworkers_f':'Non-workers'}, inplace=True)
df2.rename(columns={'nonworkers_available_f':'Non-workers available for work'}, inplace=True)

# Extract each education level and add rows corresponding to each value
dfemale_level = pd.melt(df2, id_vars=['geo_code', 'geo_level', 'area_type', 'Sex'], var_name="Workers category", value_name="Category total")
```

```python
# combine both sets
result = pd.concat([dmale_level, dfemale_level])
result = result.drop_duplicates()
result.head(5)
```

```python
output_path="/home/arun/workspace/projects/janaganana-data/data/workers_data/B-01/csv/workers_data_all_india_output.csv"
result.to_csv(output_path, index=False)
```
