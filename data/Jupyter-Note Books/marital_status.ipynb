```python
file_path="/home/arun/workspace/projects/janaganana-data/data/maritial_status/c-2/csv/marital_status.csv"

# Reading csv file for pca data
import pandas as pd
df = pd.read_csv(file_path)
```

```python
df.head(4)
```

```python
# only interested in literacy of all ages
# if age specific is required then additional conditions need to be added here
df = df[((df.total == "Rural") | (df.total == "Urban")) & (df.age_group == "All ages")]
```

```python
# pick required columns
df = df[[u'state_code', u'dist_code', u'area_name', u'total', u'age_group', u'never_married_m', u'never_married_f', u'curr_married_m', u'curr_married_f', u'widowed_m', u'widowed_f', u'separated_m', u'separated_f', u'divorced_m', u'divorced_f', u'unspecified_m', u'unspecified_f']]
```

```python
def func(ds):
    ds.area_name = ds.area_name.lower();
    if ds.area_name.startswith('district'):
        ds['geo_code'] = ds.dist_code;
    else:
        ds['geo_code'] = ds.state_code;
        
    if ds.area_name == 'india':
        ds['geo_level'] = 'country'
        ds['geo_code'] = 'IN'
    elif ds.area_name.startswith('district'):
        ds['geo_level'] = 'district'
    elif ds.area_name.startswith('state'):
        ds['geo_level'] = 'state'   
        
    return ds

df=df.apply(func, axis=1)
```

```python
df.head(4)
```

```python
df = df[[ u'geo_code', u'geo_level', u'total', u'age_group', u'never_married_m', u'never_married_f', u'curr_married_m', u'curr_married_f', u'widowed_m', u'widowed_f', u'separated_m', u'separated_f', u'divorced_m', u'divorced_f', u'unspecified_m', u'unspecified_f' ]]
```

```python
df.head(4)
```

```python
# isloate 'Male' data
df1 = df[[ u'geo_code', u'geo_level', u'total', u'never_married_m', 'curr_married_m', u'widowed_m', u'separated_m', u'divorced_m', u'unspecified_m']]
df1['Sex'] = 'Male';
df1.rename(columns={'total':'area_type'}, inplace=True)
df1.rename(columns={'never_married_m':'never married'}, inplace=True)
df1.rename(columns={'curr_married_m':'currently married'}, inplace=True)
df1.rename(columns={'widowed_m':'widowed'}, inplace=True)
df1.rename(columns={'separated_m':'separated'}, inplace=True)
df1.rename(columns={'divorced_m':'divorced'}, inplace=True)
df1.rename(columns={'unspecified_m':'unspecified'}, inplace=True)

# Extract each education level and add rows corresponding to each value
dmale_ms = pd.melt(df1, id_vars=['geo_code', 'geo_level', 'area_type', 'Sex'], var_name="marital_status", value_name="marital_status_total")
```

```python
# isloate 'Male' data
df2 = df[[ u'geo_code', u'geo_level', u'total', u'never_married_f', u'curr_married_f', u'widowed_f', u'separated_f', u'divorced_m', u'divorced_f', u'unspecified_f']]
df2['Sex'] = 'Female';
df2.rename(columns={'total':'area_type'}, inplace=True)
df2.rename(columns={'never_married_f':'never married'}, inplace=True)
df2.rename(columns={'curr_married_f':'currently married'}, inplace=True)
df2.rename(columns={'widowed_f':'widowed'}, inplace=True)
df2.rename(columns={'separated_f':'separated'}, inplace=True)
df2.rename(columns={'divorced_f':'divorced'}, inplace=True)
df2.rename(columns={'unspecified_f':'unspecified'}, inplace=True)

# Extract each education level and add rows corresponding to each value
dfemale_ms = pd.melt(df2, id_vars=['geo_code', 'geo_level', 'area_type', 'Sex'], var_name="marital_status", value_name="marital_status_total")
```

```python
# combine both sets
result = pd.concat([dmale_ms, dfemale_ms])
result = result.drop_duplicates()
```

```python
output_path="/home/arun/workspace/projects/janaganana-data/data/maritial_status/c-2/csv/marital_status_all_india_output.csv"
result.to_csv(output_path, index=False)
```
