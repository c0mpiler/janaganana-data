```python
# Reading csv file for pca data
import pandas as pd
data = pd.read_csv("/Users/mthipparthi/PycharmProjects/janaganana/data/pca_data.csv")
```

```python
# Plucking out only "Rural" and "Urban" data as "Total" can be calculated
data = data[(data.TRU == "Rural") | (data.TRU == "Urban")]
```

```python
columns_you_want = ['State','District','Level','Name','TRU', 'P_LIT', 'M_LIT', 'F_LIT', 'P_ILL', 'M_ILL', 'F_ILL'] 
data=data[columns_you_want]
data.head(5)
```

```python
#  Only taking out Male total and adding Male column to it
columns_male_lit = ['State','District','Level','Name','TRU', 'M_LIT'] 
m_lit=data[columns_male_lit]
m_lit['sex'] = 'Male'
m_lit['literacy type'] = 'Literate'
m_lit.rename(columns={'M_LIT':'Total'}, inplace=True)
m_lit.head(5)
```

```python
#  Only taking out Male total and adding Male column to it
columns_you_want = ['State','District','Level','Name','TRU', 'F_LIT'] 
female_lit=data[columns_you_want]
female_lit['sex'] = 'Female'
female_lit['literacy type'] = 'Literate'
female_lit.rename(columns={'F_LIT':'Total'}, inplace=True)
female_lit.head(5)
```

```python
#  Only taking out Male total and adding Male column to it
columns_you_want = ['State','District','Level','Name','TRU', 'M_ILL'] 
m_illlit=data[columns_you_want]
m_illlit['sex'] = 'Male'
m_illlit['literacy type'] = 'Illiterate'
m_illlit.rename(columns={'M_ILL':'Total'}, inplace=True)
m_illlit.head(5)
```

```python
columns_you_want = ['State','District','Level','Name','TRU', 'F_ILL'] 
f_illlit=data[columns_you_want]
f_illlit['sex'] = 'Female'
f_illlit['literacy type'] = 'Illiterate'
f_illlit.rename(columns={'F_ILL':'Total'}, inplace=True)
f_illlit.head(5)
```
